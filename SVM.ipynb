```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn import metrics

```

```python

# Load dataset
data = pd.read_csv(r"Social_Network_Ads.csv")

```

```python
# Select features and target
x = data.iloc[:, [2, 3]].values
y = data.iloc[:, 4].values

# Split into training and testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
```

```python

# Feature Scaling
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)

# Train SVM model
classifier = SVC(kernel='linear', random_state=0)
classifier.fit(x_train, y_train)

```

```python
# Predict
y_predict = classifier.predict(x_test)

# Accuracy
print("Accuracy Score with linear kernel:")
print(metrics.accuracy_score(y_test, y_predict))

```

```python
# Plotting test results
plt.figure(figsize=(8, 6))
plt.scatter(x_test[:, 0], x_test[:, 1], c=y_test, cmap='coolwarm')
plt.xlabel('Age (scaled)')
plt.ylabel('Salary (scaled)')
plt.title('SVM Test Data Visualization')
plt.show()

```

```python
# Plotting test results
plt.figure(figsize=(8, 6))
plt.scatter(x_test[:, 0], x_test[:, 1], c=y_test, cmap='coolwarm')
plt.xlabel('Age (scaled)')
plt.ylabel('Salary (scaled)')
plt.title('SVM Test Data Visualization')
# Plotting the decision boundary (hyperplane)
w = classifier.coef_[0]
b = classifier.intercept_[0]
a = -w[0] / w[1]
xx = np.linspace(min(x_test[:, 0])-1, max(x_test[:, 0])+1)
yy = a * xx - b / w[1]
plt.plot(xx, yy)

plt.show()
```
